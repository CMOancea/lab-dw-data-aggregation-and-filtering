```python
import pandas as pd

# Title: Loading and Cleaning Column Names
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"
df = pd.read_csv(url)

# Standardizing column names
df.columns = [col.lower().replace(" ", "_") for col in df.columns]
df.head()
```

Filtering Low Claim and Responsive Customers

```python
# Filtering Data
# Filter: total_claim_amount < 1000 and response is Yes
filtered_df = df[(df['total_claim_amount'] < 1000) & (df['response'] == 'Yes')].copy()
filtered_df.head()
```

Policy and Gender Analysis

```python
# Filter only 'Yes' responses first
yes_responses = df[df['response'].str.lower() == 'yes']

analysis = yes_responses.groupby(['policy_type', 'gender'])[['monthly_premium_auto', 'customer_lifetime_value', 'total_claim_amount']].mean()
print(analysis)
```

Customer Count by State Filter

```python
state_counts = df.groupby('state')[['customer']].count()
large_markets = state_counts[state_counts['customer'] > 500]
print(large_markets)
```

CLV Stats by Education and Gender

```python
clv_stats = df.groupby(['education', 'gender'])['customer_lifetime_value'].agg(['max', 'min', 'median'])
print(clv_stats) 
```

Bonus: Monthly Policies by State (Pivot Table)

```python
df['effective_to_date'] = pd.to_datetime(df['effective_to_date'])
df['month'] = df['effective_to_date'].dt.month_name()

pivot_table = df.pivot_table(index='state', columns='month', values='customer', aggfunc='count')
print(pivot_table)
```

Bonus: Top 3 States Performance

```python
top_3_states = df['state'].value_counts().nlargest(3).index

# 2. Filter and group
top_3_df = df[df['state'].isin(top_3_states)]
top_3_analysis = top_3_df.groupby(['state', 'month'])[['customer']].count().sort_values(by='customer', ascending=False)
print(top_3_analysis)
```

Bonus: Marketing Channel Response Rate

```python
response_rate = df.groupby('sales_channel')['response'].value_counts(normalize=True).unstack()
print(response_rate)
```
